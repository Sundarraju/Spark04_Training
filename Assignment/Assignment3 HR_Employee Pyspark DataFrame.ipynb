```
from pyspark.sql.functions import *
```

a)	Create HR_Employee Pysaprk dataFrame

```
HR_Employee_df = spark.read.csv("dbfs:/FileStore/tables/HR_Employee.csv", header = True, inferSchema = True)
type(HR_Employee_df)
```

b)	Show printSchema & show 10 rows of Dataset

```
HR_Employee_df.printSchema()
HR_Employee_df.show(10)
```

c)	Most employee is working in which department

```
HR_Employee_df.createOrReplaceTempView("HR_emp_Data")

spark.sql("select Department,count(EmployeeID) as Total from HR_emp_Data group by Department order by Total desc limit 1").show()
```

d)	Highest number of job roles

```
spark.sql("select JobRole,count(JobRole) as number from HR_emp_Data group by JobRole order by number desc limit 1").show()
```

e)	Which gender have higher strength as workforce?

```
spark.sql("select Gender, count(Gender) as number from HR_emp_Data group by Gender order by number desc limit 1").show()
```

f)	Compare the marital status of employee and find the most frequent status.

```
spark.sql("select MaritalStatus, count(MaritalStatus) as number from HR_emp_Data group by MaritalStatus order by number desc ").show()
```

g)	Mostly hired employee have qualification

```
spark.sql("select Education, count(Education) as number from HR_emp_Data group by Education order by number desc").show()
```

h)	Find the count of employee from which education fields

```
spark.sql("select EducationField,count(EducationField) as number from HR_emp_Data group by EducationField order by number desc").show()
```

i)	What is the job satisfaction level of employee?

```
spark.sql("select JobSatisfaction, count(JobSatisfaction) as number from HR_emp_Data group by JobSatisfaction order by number").show()
```

j)	Does most of employee do overtime: Yes or No?

```
spark.sql("select OverTime, count(OverTime) as total from HR_emp_Data group by OverTime order by total desc ").show()
```

k)	Find Min & Max Salaried employees.

```
HR_Employee_df.select(max('Income')).show() 
HR_Employee_df.select(min('Income')).show()
```

l)	Does most of the employee do business travel? Find of the employees counts for each category

```
spark.sql("select BusinessTravel,count(BusinessTravel) as Emp_count from HR_emp_Data group by BusinessTravel order by Emp_count desc ").show()
```

m)	Find the AVG salary of graduate employee.

```
HR_Employee_df.select(avg(when(HR_Employee_df['Education'] == 'Bachelor',HR_Employee_df['Income'])).alias("avg_Income_grad")).show()
```
